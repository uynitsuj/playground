```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
filename = 'data/kc_house_data.csv'
data = np.genfromtxt(filename, delimiter=",", usecols=(2, 5), names=True)
fig, ax = plt.subplots()
ax.plot(data['sqft_living'],data['price'], 'o', markersize = 1)
plt.title("King County, USA Housing Prices")
plt.xlabel("Living Area ($m^2$)")
plt.ylabel("Price ($)")
```

```python
sumx = np.sum(data['sqft_living'])
sumy = np.sum(data['price'])
sumx_sq = np.sum(np.square(data['sqft_living']))
sumxy = np.sum(np.multiply(data['sqft_living'],data['price']))
N = len(data['sqft_living'])
m = (N*sumxy-sumx*sumy)/(N*sumx_sq-sumx**2)
b = (sumy-m*sumx)/N
x = np.linspace(0,14000,14000)
y = m*x+b
ax.plot(x,y)
fig
```
