```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
filename = 'data/kc_house_data.csv'
data = np.genfromtxt(filename, delimiter=",", usecols=(2, 5), names=True)
plt.plot(data['sqft_living'],data['price'], 'o')
plt.title("King County, USA Housing Prices")
plt.xlabel("Living Area ($m^2$)")
plt.ylabel("Price ($)")
```
